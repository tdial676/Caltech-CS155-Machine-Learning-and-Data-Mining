```python
import numpy as np
from matplotlib import pyplot as plt
from numpy import genfromtxt
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
%matplotlib inline
```

## Load data

```python
import urllib.request
urllib.request.urlretrieve('https://raw.githubusercontent.com/emiletimothy/Caltech-CS155-2023/main/set2/data/problem3data.txt', 'problem3data.txt')

train_data = np.loadtxt('problem3data.txt')
y_train = train_data[:, 9]
X_train = train_data[:, :9]
```

```python
from os import RTLD_LOCAL
l_alpha = np.linspace(0.01, 3, 20)
r_alpha = [i for i in range(1, 10001)]
l_weight = []
r_weight = []

for l in l_alpha:
  l_clf = Lasso(alpha = l)
  l_clf.fit(X_train, y_train)
  l_weight.append(l_clf.coef_)


for r in r_alpha:
  r_clf = Ridge(alpha = r)
  r_clf.fit(X_train, y_train)
  r_weight.append(r_clf.coef_)

plt.figure()
for w in np.array(l_weight).T:
 plt.plot(l_alpha, w, marker = 'o', linewidth = 1)
plt.figure()
for w in np.array(r_weight).T:
  plt.plot(r_alpha, w)
```
